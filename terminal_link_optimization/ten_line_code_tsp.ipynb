```python
# Vanilla TSP solution with starting point at 0,0

import random, numpy, math, copy, matplotlib.pyplot as plt
import numpy as np

# Define the distance between ground links and satellite here
cities = [random.sample(range(100), 2) for x in range(15)];
# Define starting node
cities[0] = [0,0]
# Define number of ground links
tour = random.sample(range(15),15);
for temperature in numpy.logspace(0,5,num=100000)[::-1]:
    # Choose a random city in list
    [i,j] = sorted(random.sample(range(15),2));
    newTour =  tour[:i] + tour[j:j+1] +  tour[i+1:j] + tour[i:i+1] + tour[j+1:];
    if math.exp( ( sum([ math.sqrt(sum([(cities[tour[(k+1) % 15]][d] - cities[tour[k % 15]][d])**2 for d in [0,1] ])) for k in [j,j-1,i,i-1]]) - sum([math.sqrt(sum([(cities[newTour[(k+1) % 15]][d] - cities[newTour[k % 15]][d])**2 for d in [0,1] ])) for k in [j,j-1,i,i-1]])) / temperature) > random.random():
        tour = copy.copy(newTour);
plt.plot([cities[tour[i % 15]][0] for i in range(16)], [cities[tour[i % 15]][1] for i in range(16)], 'xb-');
plt.show()
```

```python
# CubeSat version of TSP
"""
Essentially make it such that the salesman must travel 
back to the starting point (CubeSat) after each trip.
Not sure if this is the most optimal way to do things, 
but it's a start ...
"""
import random, numpy, math, copy, matplotlib.pyplot as plt
import numpy as np

# Define the distance between ground links and satellite here
cities = [random.sample(range(100), 2) for x in range(15)];
# Define location of satellite
cities[0] = [0,0]
# Define number of ground links + satellite
tour = random.sample(range(15),15);
for temperature in numpy.logspace(0,5,num=100000)[::-1]:
    # Choose a random city in list
    [i,j] = sorted(random.sample(range(15),2));
    newTour =  tour[:i] + tour[j:j+1] +  tour[i+1:j] + tour[i:i+1] + tour[j+1:];
    if math.exp( ( sum([ math.sqrt(sum([((cities[0][d] - cities[tour[k % 15]][d])**2) for d in [0,1] ])) for k in [j,j-1,i,i-1]]) - sum([math.sqrt(sum([((cities[0][d] - cities[newTour[k % 15]][d])**2) for d in [0,1] ])) for k in [j,j-1,i,i-1]])) / temperature) > random.random():
        tour = copy.copy(newTour);
plt.plot([cities[tour[i % 15]][0] for i in range(16)], [cities[tour[i % 15]][1] for i in range(16)], 'xb-');
plt.show()
```
